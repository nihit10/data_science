```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [i for i in np.arange(0, 10, 0.1)]
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y)
```

```python
_y = np.cos(x)
```

```python
plt.figure(figsize=(15,10))
plt.plot(x, _y)
plt.legend("x")
plt.plot(x,y)
plt.legend("Cos")
```
