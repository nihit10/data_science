```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.naive_bayes import BernoulliNB
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
np.random.seed(1000)
nb_samples = 300
```

```python
def visual_dataset(X,Y):
    fig, ax = plt.subplots(1,1, figsize=(30,25))
    
    ax.grid()
    ax.set_xlabel('X')
    ax.set_ylabel("Y")
    
    for i in range(nb_samples):
        if Y[i] ==0:
            ax.scatter(X[i, 0], X[i,1], marker="o", color="r")
        else:
            ax.scatter(X[i, 0], X[i,1], marker="^", color="b")
    plt.show()
```

```python
X, Y= make_classification(n_samples=nb_samples, n_features=2,
                          n_informative=2, n_redundant=0)
visual_dataset(X,Y)
```

```python

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25)
```

```python
bnb = BernoulliNB(binarize=0.0)
```

```python
bnb.fit(X_train, Y_train)
```

```python
print("Bernoulli Naive Bayes Score %.3f" % bnb.score(X_test, Y_test))
```

```python
#Compute cross validation score
bnb_scores = cross_val_score(bnb, X, Y, scoring='accuracy', cv=10)
print("Berboulli Naive Bayes CV average score: %.3f" % bnb_scores.mean())
```

```python
#checking naive Bayes Predictions, we obtain:
data = np.array([[0,0], [0,1],[1,0],[1,1],[1, 0]])
```

```python
bnb.predict(data)
```
